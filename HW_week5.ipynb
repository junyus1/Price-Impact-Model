### Q1

```q
computeImpact: {[trade; adv; vol; h] 
                beta: (log 2) % (h*60);
                lamda: 1;
                dt: 10 /here is the time interval
                dvol: vol * sqrt 2340;
                :(neg count trade)#ema[beta * dt; (0, lamda * dvol * trade % (beta * dt * adv))];}
```

```q
tbl_test: select from tbl where id = 0
trade_test: tbl_test`trade 
adv_test: tbl_test`adv
vol_test: tbl_test`vol
h: 1
computeImpact[trade_test;adv_test;vol_test;h]
```

### Q4

```q
dt: 1, 15, 60
tbl4: ([]dt:dt) cross tbl3

tbl4: update Return: (neg 1) + xprev[-6*dt; mid] % mid,
             dI: xprev[-6*dt; Impact] + (neg Impact) 
             by dt, h, id from tbl4;

cols tbl4
tbl4
```

```q
select Return: Return
       by id, h from tbl4
```

```q
select dI: dI
       by id, h from tbl4
```

### Q5
